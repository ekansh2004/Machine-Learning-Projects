```python
import os
import shutil
import string
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from google.colab import drive
pd.set_option('display.max_colwidth', 1)
```

```python
drive.mount('/content/drive')
file_path = '/content/drive/My Drive/cleaned_train.csv'
```

```python
train = pd.read_csv(file_path)
```

```python
train.head()
```

```python
def createFeature(df):
    return [x.split(' ')[y] for x,y in zip(df['TEXT'], df['LOCATION'])]
```

```python
train['ABV'] = createFeature(train)
```

```python
def dropCols(df):
    return df.drop(columns=['ABSTRACT_ID', 'LOCATION', 'TEXT'])
```

```python
train = dropCols(train)
```

```python
train.head()
```

```python
save_path = '/content/drive/My Drive/final_train.csv'
train.to_csv(save_path, index=False)
print('final_train saved')
```
